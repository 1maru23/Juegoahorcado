**Question 2:** In how many different states do the customers live in?

```python
import csv

datafile = "./sample_data/customers.csv"
states = set()

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl)
    header = next(csvreader)

    for row in csvreader:
        if len(row) > 4:
            state = row[4].strip()
            if state != "":
                states.add(state)

print("Estados únicos:", len(states))
print(states)
```

**Question 3** What is the state with most customers?

```python
import csv

datafile = "./sample_data/customers.csv"
state_count = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl)
    header = next(csvreader)

    for row in csvreader:
        if len(row) > 4:
            state = row[4].strip()

            if state != "":
                state_count[state] = state_count.get(state, 0) + 1

# obtener estado con más clientes
max_state = max(state_count, key=state_count.get)

print("Estado con más clientes:", max_state)
print("Cantidad:", state_count[max_state])
```

```python
import csv

datafile = "./sample_data/customers.csv"
state_count = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl)
    header = next(csvreader)

    for row in csvreader:
        if len(row) > 4:
            state = row[4].strip()

            if state != "":
                state_count[state] = state_count.get(state, 0) + 1

# encontrar el valor mínimo
min_value = min(state_count.values())

# encontrar TODOS los estados con ese valor
min_states = [state for state, count in state_count.items() if count == min_value]

print("Estados con menos clientes:", min_states)
print("Cantidad:", min_value)
```

```python
import csv

datafile = "./sample_data/orders.csv"
rows = []

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl)
    header = next(csvreader)

    for row in csvreader:
        if len(row) >= 6:
            rows.append([col.strip() for col in row])

print("Filas cargadas:", len(rows))
print("Columnas:", header)
print("Primera fila:", rows[0])
```

```python
unique_orders = set()

for row in rows:
    order_id = row[1]
    if order_id != "":
        unique_orders.add(order_id)

print("Órdenes únicas:", len(unique_orders))
```

```python
order_items = {}

for row in rows:
    order_id = row[1]
    if order_id != "":
        order_items[order_id] = order_items.get(order_id, 0) + 1

average_items = sum(order_items.values()) / len(order_items)
print("Promedio de items por orden:", round(average_items, 2))
```

```python
max_items = max(order_items.values())
print("Máximo de items en una orden:", max_items)
```

```python
october_orders = set()

for row in rows:
    order_id = row[1]
    date = row[2]

    if order_id != "" and date.startswith("2021-10"):
        october_orders.add(order_id)

print("Órdenes en octubre 2021:", len(october_orders))
```

```python
import csv
customers_file = "./sample_data/customers.csv"
customer_names = {}

with open(customers_file, 'r') as fl:
    csvreader = csv.reader(fl)
    header = next(csvreader)

    for row in csvreader:
        if len(row) >= 3:
            customer_id = row[0].strip()
            first_name = row[1].strip()
            last_name = row[2].strip()

            if customer_id != "":
                customer_names[customer_id] = f"{first_name} {last_name}"
orders_file = "./sample_data/orders.csv"
customer_spend_2021 = {}
processed_orders = set()

with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl)
    header = next(csvreader)

    for row in csvreader:
        if len(row) >= 4:
            customer_id = row[0].strip()
            order_id = row[1].strip()
            date = row[2].strip()
            order_total = row[3].strip()

            if (
                customer_id != "" and
                order_id != "" and
                date.startswith("2021") and
                order_id not in processed_orders
            ):
                try:
                    total = float(order_total)
                    customer_spend_2021[customer_id] = customer_spend_2021.get(customer_id, 0) + total
                    processed_orders.add(order_id)
                except:
                    pass

top_customer = max(customer_spend_2021, key=customer_spend_2021.get)
top_name = customer_names.get(top_customer, "Desconocido")

print("Cliente que más gastó en 2021:")
print("ID:", top_customer)
print("Nombre:", top_name)
print("Total:", round(customer_spend_2021[top_customer], 2))
```

```python
monthly_sales = {}
processed_orders = set()

for row in rows:
    order_id = row[1]
    date = row[2]
    order_total = row[3]

    if order_id != "" and date != "" and order_id not in processed_orders:
        try:
            total = float(order_total)
            month = date[:7]   # formato YYYY-MM
            monthly_sales[month] = monthly_sales.get(month, 0) + total
            processed_orders.add(order_id)
        except:
            pass

best_month = max(monthly_sales, key=monthly_sales.get)
print("Mejor mes histórico:", best_month)
print("Ventas:", round(monthly_sales[best_month], 2))
```
